```python
import pandas as pd
import numpy as np
import numpy as np
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import StratifiedKFold
```

```python
#10 folds para validação cruzada
k_folds = 10
```

```python
"""Pegando os dados de treinamento de fac - CRISP"""
X_fac_original = pd.read_csv('datasets/mfeat-fac-crisp.csv', delimiter =',', header=None)
X_fac_original.head()
```

```python
#Pega todas as linhas de todos os exemplos até a coluna 216 (exclusive) . Na prática (o que se quer) pega até a coluna 215
#Esses são nossos dados sem a rotulação de classes (label)
X_fac_train = X_fac_original.iloc[:, 0: 216].get_values() # retorna como uma matriz e não como dataframe 
X_fac_train
```

```python
#Pega a última coluna com todas suas linhas. Isso será nosso array de true label, com a marcação correta das classes
y_labels = X_fac_original.iloc[:, 216].get_values()
y_labels
```

```python
#K= 10.. OU SEJA: 10 FOLDS
skf = StratifiedKFold(n_splits=k_folds)
```

```python
#faz a divisão preservando a percentagem igual de classes por fold
skf.get_n_splits(X_fac_train, y_labels)
```

```python
#Naive Bayes Classifier
clf = GaussianNB()
```

```python
somatorio_acuracia = 0
```

```python
for train_index, test_index in skf.split(X_fac_train, y_labels):
    
    """
for de i a K
	pega um fold para validação e o restante para treinamento (o método que faz isso por si)
	treina o modelo com os folds de treinamento e testa com os folds de validação pra ver se a predições estão corretas
	pega a acurácia

pega a média das acurácias"""
    
for train_index, test_index in skf.split(X_fac_train, y_labels):

    """Separando os dados para treino e teste
    X_train, X_test = X_fac_train[train_index], X_fac_train[test_index]
    
    # Separando os folds LABELED para conferir com o treino (y_train) E verificar a acurácia do teste (y_test)
    y_train, y_test = y_labels[train_index], y_labels[test_index]
      
    #Treinando.......................................
    #Treinando o modelo GaussianNB: associando os folds correspondentes para treinamento, OU SEJA, 
    #dados de treino x com os true labels de y para TREINAR
    clf.fit(X_train, y_train)
    
    3Testando.......................................
    
    #Obtendo a acurácia atual: esse método retorna a média de acerto DESSA rodada de testes
    #score: Returns the mean accuracy on the given test data and labels.
    acuracia_atual = clf.score(X_test, y_test)
    
    #Fazendo somatório das acurácias para posterior média
    somatorio_acuracia += acuracia_atual
   
   # print("acuracia_atual:", acuracia_atual)     """
```

```python
media_acuracia = somatorio_acuracia / k_folds
media_acuracia
```
