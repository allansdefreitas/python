```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from sklearn.linear_model import LinearRegression
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

```python
x = np.array([[18], [20], [25], [22], [27], [31], [33], [39], [47], [58]])
```

```python
x
```

```python
y = np.array([[871], [920], [1100], [1234], [1360], [1450], [1530], [1702], [1876], [1982]])
```

```python
y
```

```python
plt.scatter(x,y)
```

```python
regressor = LinearRegression()
regressor.fit(x, y)
```

```python
# b0 constant
regressor.intercept_
```

```python
#b1 coefficient
regressor.coef_
```

```python
prevision1 = regressor.intercept_ + regressor.coef_ * 40
prevision1
```

```python
#does not work!!
prevision2 = regressor.predict(40)
prevision2
```

```python
previsions = regressor.predict(x)
previsions
```

```python
meanSquaredError = (abs(y - previsions))
meanSquaredError
```

```python
meanSquaredError = MeanSquaredError.mean()
meanSquaredError
```

```python
# to test
mae = mean_absolute_error(y, previsions)

# to train
mse = mean_squared_error(y, previsions)
```

```python
mae
```

```python
mse
```

```python
plt.plot(x,y, 'o')
plt.plot(x, previsions, color = 'red')
plt.title('Simple Linear Regression')
plt.xlabel('Age')
plt.ylabel('Custo ($)')
```
