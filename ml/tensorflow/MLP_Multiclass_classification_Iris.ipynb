```python
neurons_input_layer = X.shape[1] #
neurons_input_layer
```

```python
#Formula to discover the number of neurons in hidden layer: (number of atts PLUS number of classes) DIVIDED BY 2
neurons_hidden_layer = int(np.ceil((X.shape[1] + y.shape[1]) / 2))
neurons_hidden_layer
```

```python
neurons_exit_layer = y.shape[1]
neurons_exit_layer
```

```python
W = {'hidden': tf.Variable(tf.random_normal([neurons_input_layer, neurons_hidden_layer])),
     'exit': tf.Variable(tf.random_normal([neurons_hidden_layer, neurons_exit_layer]))}
```

```python
b = {'hidden': tf.Variable(tf.random_normal([neurons_hidden_layer])),
     'exit': tf.Variable(tf.random_normal([neurons_exit_layer]))}
```

```python
#Using placeholders: 
xph = tf.placeholder('float', [None, neurons_input_layer])
yph = tf.placeholder('float', [None, neurons_exit_layer])
```

```python
def mlp(X, W, bias):
    hidden_layer = tf.add( tf.matmul( X, W['hidden']), bias['hidden'] )
    hidden_layer_activation = tf.nn.relu(hidden_layer)
    
    exit_layer = tf.add( tf.matmul( hidden_layer_activation, W['exit']), bias['exit'] )
    return exit_layer
```

```python
model = mlp(xph, W, b)
```
